# <font style ="color:#ff2e54; font-size:24px" >Pie Chart</font>

```python
plt.figure(figsize=(7,7))
l=df['Classification'].value_counts()
plt.pie(l,labels=l.index,shadow=True, textprops={'fontsize': 14},autopct='%1.2f%%')
plt.legend(fontsize=10,loc="best")
plt.title("Classification - Pie Chart",fontsize=20)
plt.show()
```

# <font style ="color:#ff2e54; font-size:24px" >Lineplot - Sepal Width vs Sepal Length</font>

```python
#Lineplot - Sepal Width vs Sepal Length

plt.figure(figsize=(15,7))
sns.lineplot(x = "SW", y ="SL",hue = "Classification", data = df)

plt.show()
```

# <font style ="color:#ff2e54; font-size:24px" >Lineplot - Petal Width vs Petal Length</font>

```python
#Lineplot - Petal Width vs Petal Length

plt.figure(figsize=(15,7))
sns.lineplot(x = "PW", y ="PL",hue = "Classification", data = df)

plt.show()
```
